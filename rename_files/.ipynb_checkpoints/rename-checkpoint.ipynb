## Copy files

```python
base = "/Volumes/NO NAME/La_Belle_Sauvage"
target = "/Volumes/NO NAME/La_Belle_Sauvage_2"

filelist = os.listdir(base)
filelist.sort()
#filelist

for f in filelist:
    file_old = base + '/' + f
    file_new = target + '/' + f
    copyfile(file_old,file_new)
```

```python
base = "/Volumes/NO NAME/Factfulness"
target = "/Volumes/NO NAME/Factfulness_2"

filelist = os.listdir(base)
filelist.sort()
#filelist

for f in filelist:
    file_old = base + '/' + f
    file_new = target + '/' + f
    copyfile(file_old,file_new)
    print(f)
```
