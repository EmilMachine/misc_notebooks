# 1.3 Baysian inference

$ p(A|B) $ = Conditional dist of A given B

$p(\cdot)$ is distribution (full graph)

$Pr(\cdot)$ is probablity (specific integration)


$ \widetilde{y} $ = (yet) unseen outcomes

$p( \theta | y ) $  or  $p( \widetilde{y} | y )$  = "future" based on observed outcomes



## 1.3.1 

$p(\theta \cap y) =p(\theta,y) =p(\theta)p(y|\theta)$

**Baye's rule**:
$p(\theta|y) = \frac{p(\theta)p(y|\theta)}{p(y)} $     
---
prior predictive distribution / marginal distribution

$p(y)=\int p(\theta)p(y|\theta)d\theta $

posterior preditive distribution

$p( \widetilde{y} | y ) = \int p(\widetilde{y} | \theta) p(\theta | y) $

## Exercise 1.1

 Conditional probability: suppose that if θ = 1, then y has a normal distribution with mean 1 and standard deviation σ, and if θ = 2, then y has a normal distribution with mean 2 and standard deviation σ. Also, suppose Pr(θ = 1) = 0.5 and Pr(θ = 2) = 0.5.
- (a) For σ = 2, write the formula for the marginal probability density for y and sketch it.
- (b) What is Pr(θ = 1|y = 1), again supposing σ = 2?
- (c) Describe how the posterior density of θ changes in shape as σ is increased and as it is decreased.


---
a) marginal distribution of y

$p(y) = \int{p(y,\theta)d\theta} = \int{p(\theta) p(y|\theta)d\theta}$ 

$p(y) = p(\theta=1)p(y|\theta=1) +  p(\theta=2)p(y|\theta=2)$
$ = 1/2 \cdot \left(p(y|\theta=1)+p(y|\theta=2) \right)$


```python
# Pr(θ = 1) = 0.5 and Pr(θ = 2) = 0.5. , σ = 2
# marginal probability density for y 

import scipy.stats as ss
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('ggplot')

# font = {'weight' : 'bold',
#         'size'   : 22}
# mpl.rc('font', **font)
#mpl.rcParams['font.family'] = "serif"
# plt.rcParams.update(
#                         {'font.size': 16
#                  ,'figure.figsize':(6.0, 4.0)}  
#                        )


%matplotlib inline
```

```python
x = np.arange(-5,9,0.1)
y = 0.5*(ss.norm.pdf(x=x, loc=1, scale=2) + ss.norm.pdf(x=x, loc=2, scale=2))
plt.figure(figsize=(8, 2))
plt.plot(x,y)
plt.title("marginal distribution of y, $\\theta=[1,2]$")
plt.xlabel("x",fontsize=10)
plt.ylabel("y");
ax=plt.gca()
minor_locator   = mpl.ticker.AutoMinorLocator(2)
ax.xaxis.set_minor_locator(minor_locator)
plt.grid(which='minor')
plt.show()
```

```python
x = np.arange(-5,18,0.1)
y = 0.5*(ss.norm.pdf(x=x, loc=1, scale=2) + ss.norm.pdf(x=x, loc=10, scale=2))
plt.figure(figsize=(8, 2))
plt.plot(x,y)
plt.title("marginal distribution of y, $\\theta=[1,10]$")
plt.xlabel("x")
plt.ylabel("y");
ax=plt.gca()
minor_locator   = mpl.ticker.AutoMinorLocator(2)
ax.xaxis.set_minor_locator(minor_locator)
plt.grid(which='minor')
```

b) What is Pr(θ = 1|y = 1), again supposing σ = 2?

$$ p(\theta|y) = \frac{p(\theta)p(y|\theta)}{p(y)} $$

$$ Pr(\theta=1|y=1) = \frac{Pr(\theta=1)Pr(y=1|\theta=1)}{Pr(\theta=1)Pr(y=1|\theta=1) + Pr(\theta=2)Pr(y=1|\theta=2)}$$

$$ =\frac{0.5\cdot0.199}{0.5\cdot0.199+0.5\cdot0.176} = 0.53$$

```python


sigma=2
pr1 = ss.norm.pdf(x=1, loc=1, scale=sigma)
pr2 = ss.norm.pdf(x=1, loc=2, scale=sigma)

pr1given1 = pr1/(pr1+pr2)

print(f"pr1:{pr1}\npr2:{pr2}\npr1|1:{pr1given1}")
```

```python
x = np.arange(-1,3,0.1)
y1 = 0.5*ss.norm.pdf(x=x, loc=1, scale=2)
y2 = 0.5*ss.norm.pdf(x=x, loc=2, scale=2)
plt.figure(figsize=(4, 2))
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["$\\mu=1$","$\\mu=2$"])
plt.title("contribution, $\\theta=[1,2]$")
plt.xlabel("x",fontsize=10)
plt.ylabel('"$y_i$"');
ax=plt.gca()
minor_locator   = mpl.ticker.AutoMinorLocator(2)
ax.xaxis.set_minor_locator(minor_locator)
plt.grid(which='minor')
plt.show()
```

```python
sigma=np.arange(0.01,10,0.01)

def calc(sigma):

    pr1 = ss.norm.pdf(x=1, loc=1, scale=sigma)
    pr2 = ss.norm.pdf(x=1, loc=2, scale=sigma)
    pr1given1 = pr1/(pr1+pr2)
    return pr1given1

prob = list(map(calc,sigma))

plt.figure(figsize=[3,2],dpi=150)
plt.plot(sigma,prob)
plt.ylim([0,1.1])
plt.xlabel("$\\sigma$")
plt.ylabel("p($\\theta$=1|y=1)")
```

```python
def myplot(sigma):

    x = np.arange(-1,3,0.01)
    y1 = 0.5*ss.norm.pdf(x=x, loc=1, scale=sigma)
    y2 = 0.5*ss.norm.pdf(x=x, loc=2, scale=sigma)
    plt.figure(figsize=(4, 2))
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.legend(["$\\mu=1$","$\\mu=2$"])
    plt.title(f"contribution, $\\theta=[1,2]$,$\\sigma$={sigma}")
    plt.xlabel("x",fontsize=10)
    plt.ylabel('"$y_i$"');
    ax=plt.gca()
    minor_locator   = mpl.ticker.AutoMinorLocator(2)
    ax.xaxis.set_minor_locator(minor_locator)
    plt.grid(which='minor')
    plt.show()
```

```python
myplot(0.1)
```
