```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
p = 0.5
n_jumps = 1000
n_trials = 2000

jumps = np.random.binomial(1,p,(n_jumps,n_trials))*2-1

print(np.shape(jumps))

traces = np.cumsum(jumps,0)
```

```python
plt.plot(traces[:,:10],alpha=0.5)
plt.plot([0,500],[0,0])
```

```python
# how many touches zero

n_last_jumps = 500


zero_equal = np.sum(traces[-n_last_jumps:,:]==0,0)

has_negative = (np.sum(traces[-n_last_jumps:,:]<0,0)>0).astype(int)
has_positive = (np.sum(traces[-n_last_jumps:,:]>0,0)>0).astype(int)

zero_crossing = (has_negative+has_positive ==2)


touch_zero = np.mean((zero_equal>0))
avg_touch_zero = np.mean(zero_equal)

cross_zero = np.mean(zero_crossing)


print(f""""
==========================
Last {n_last_jumps} of {n_jumps} jumps
==========================
touch_zero: {touch_zero}
cross_zero: {cross_zero}
avg_touch_zero: {avg_touch_zero}
""")
```

```python
plt.hist(zero_equal,bins=20);
plt.title("touch zero distribution")
plt.xlabel("number of times trace = 0")
plt.ylabel("number of traces")
```
