```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import imread
from scipy.spatial import cKDTree
import random

```

```python
# function to transform color image to grayscale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

def rgb2hex(color):
    '''
    Matplotlib scatter is not happy with rgb tuples so we need to transform them to hex
    '''
    c = tuple([np.int(255 if c == 1.0 else c * 256.0) for c in color])
    return "#%02x%02x%02x" % c

# function to generate random colors
# gen colors 
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
# parameters
p = 0.01 # propability of selecting a pixel/node
k = 15 # number of connections pre per pixel/node
pix_threshold = 0.9  # remove values above this value 0 (white) - 255 (black) OR 0 (black) - 1 (white)
```

```python
# load image
#data = plt.imread('in/atlas.png')
data = plt.imread('data/xy.png')
y,x = np.where(rgb2gray(data[:,:,:3])<pix_threshold)
y_norm, x_norm = map(float,data[:,:,0].shape)
colors = data[:,:,:3]

# if its a large image it might be a good idea to downsample
# y,x = np.where(rgb2gray(data[::3,::3,:3])<pix_threshold)
# y_norm, x_norm = map(float,data[::3,::3,0].shape)
# colors = data[::3,::3,:3]
```

```python
plt.imshow(data)
```

```python
# select nodes
X = np.array(random.sample(list(zip(x,y)),int(len(y)*p)))*1.0

# find k nearest neighbors using scipy.spatial.cKDTree
tree = cKDTree(X)
```

```python
# construct figure
plt.figure(figsize=(x_norm/120.,y_norm/120.))
ax = plt.subplot(111)

# create lists for position of links
x_ = []
y_ = []

# go through each node and construct links 
for pt in X:
    # find k nearest neighbors
    dist, ind = tree.query(pt, k=k+1) # k' = k+1 because method returns points itself 
    for kneigh in ind[1:]:
        x_.append([pt[0],X[kneigh][0]])
        y_.append([pt[1],X[kneigh][1]])

        
mode = "colororder"

if (mode == "gray"):
    plt.plot(np.array(x_).T,np.array(y_).T,color='#282828',lw=0.8,alpha=0.4,zorder=2)

elif(mode[:5] =="color"):
    # Plot colors
    np.random.seed(42)
    npoints = np.size(X,0)
    ncolors = 100

    alphalist = np.random.rand(npoints)/3.0
    
    # color random
    colorseed = np.random.randint(0,100,npoints)
   
    if (mode=="colororder"):
        # horizental
        colorseed = (ncolors*(y-min(y))/(max(y)-min(y))).astype(int)
        # vertical
        colorseed = (ncolors*(x-min(x))/(max(x)-min(x))).astype(int)
    
    colorfunc = get_cmap(ncolors,"gnuplot")
    colorlist = list(map(colorfunc,colorseed)) 

    if (mode=="colororiginal"):
        colorlist = [rgb2hex(colors[int(xx),int(yy),:]) for yy,xx in X] # colors

    
    for i,c in enumerate(colorlist):
        plt.plot(np.array(x_).T[:,i*k:(i+1)*k-1]
             ,np.array(y_).T[:,i*k:(i+1)*k-1]
             ,color=c
             ,lw=0.8
             ,alpha=alphalist[i]
             ,zorder=2) 



# unpack nodes
y,x = zip(*X)

# plot using a single color
#plt.scatter(y,x,marker='o',c='#282828',s=0.5,alpha=1)

# or if you want to draw the network with the original colors of your image
# c = [rgb2hex(colors[int(xx),int(yy),:]) for yy,xx in X] # colors
# plt.scatter(y,x,marker='o',c=c,s=3,alpha=1,zorder=3)

plt.axis('off')
plt.ylim(y_norm,0)
plt.xlim(0,x_norm)
plt.tight_layout()
plt.savefig('data/xy_network.png',dpi=250,pad=0.0,bbox_inches='tight')
#plt.close()
```
